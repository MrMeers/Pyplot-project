```python
import pandas as pd
import os
from matplotlib import pyplot as plt
import numpy as np
os.getcwd()
```

```python
yt=pd.read_csv('USvideos.csv')
```

```python
mkbhd=yt[yt['channel_title']=='Marques Brownlee']
```

```python
mkbhd1=mkbhd.reset_index().sort_values('views')
plt.style.use('seaborn')
plt.barh(mkbhd1['title'], mkbhd1['views'])
plt.title('MKBHD Trending Videos 2017-18')
```

```python
import plotly.express as px
px.bar(mkbhd1, mkbhd1['title'], mkbhd1['views'],
             hover_data=[mkbhd1['title'], mkbhd1['views']], color='views',
             labels=mkbhd1['title'], height=400)
```
